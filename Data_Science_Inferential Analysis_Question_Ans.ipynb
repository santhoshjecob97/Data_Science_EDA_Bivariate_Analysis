## 1)Replace the NaN values with correct value. And justify why you have chosen the same.

Since these individuals did not earn a salary, imputing these missing values with the mean or median would be inappropriate. Therefore, to accurately represent their financial outcome, we will replace the null salary figures with zero for data intergrity purpose. We use zero because:
1. It accurately represents that unplaced candidates have no salary (0)
2. Using mean/median would introduce artificial values that don't reflect reality
3. Zero maintains mathematical integrity for future calculations
4. It's more meaningful than leaving values as NaN
- **Step 4 (optional) : Checking the missing values in categorical columns using to fill the mode values (most frequent values) or fill with unknown values.

## 7)Test the Analysis of Variance between etest_p and mba_p at signifance level 5%.(Make decision using Hypothesis Testing)

```python
from scipy.stats import f_oneway
import scipy.stats as stats

f_statistic, p_value = stats.f_oneway(dataset['etest_p'],dataset['mba_p'])

print(f"F-statistic: {f_statistic}")
print(f"P-value: {p_value}")
```

- **H0 (null hypothesis): There is no significant difference between columns
- **H1 (alternative hypothesis): There is significant difference between columns
- **Result : P(6.72 - 16)<0.05 or 5%, So H1 (alternative hypothesis): There is significant difference between columns.


### Decision Rule (α = 0.05):
- If p-value ≤ 0.05 → Reject H₀ (significant difference exists).
- If p-value > 0.05 → Fail to reject H₀ (no significant difference).So choose H1.


### Interpretation:
- There is strong evidence (p ≈ 0) campare to (p ≈ 0.05) is low,  that etest_p and mba_p have statistically different means(H1).
- There is a case         (p ≈ 0) campare to (p ≈ 0.05) is high,  that etest_p and mba_p have statistically different means(H0).

## 8)Test the similarity between the degree_t(Sci&Tech) and specialisation(Mkt&HR) with respect to salary  at significance level of 5%.(Make decision using Hypothesis Testing)

#### Independant Sample- Unpaired T Test
Diferrent Group(degree_t, specialisation), but same contion(salary)

```python
from scipy.stats import ttest_ind

# Filter data
sci_tech = dataset[dataset['degree_t'] == 'Sci&Tech']['salary'].dropna()
mkt_hr = dataset[dataset['specialisation'] == 'Mkt&HR']['salary'].dropna()

# Perform t-test
t_stat, p_value = ttest_ind(sci_tech, mkt_hr, equal_var=False)  
print(f"t-statistic: {t_stat:.4f}, p-value: {p_value:.4f}")
```

- **H0 (null hypothesis): There is no significant difference between columns
- **H1 (alternative hypothesis): There is significant difference between columns
- **Result : P(0.011)<0.05 or 5%, So H1 (alternative hypothesis): There is significant difference between columns.

## 9)Convert the normal distribution to standard normal distribution for salary column
- Why This Matters:

- Avoids Bias: Ignoring NaN values during calculations prevents skewed statistics.

- Flexibility: Lets you decide later how to handle missing values (e.g., drop or impute).

- Safety: Prevents errors from propagating NaN through arithmetic operations

```python
def stdNBgraph(dataset):
        # Coverted to standard Normal Distribution
    import seaborn as sns
    mean=dataset.mean()
    std=dataset.std()

    values=[i for i in dataset]

    z_score=[((j-mean)/std) for j in values]

    sns.distplot(z_score,kde=True)

    sum(z_score)/len(z_score)
    #z_score.std()
```

```python
stdNBgraph(dataset['salary'])
```

## 10)What is the probability Density Function of the salary range from 700000 to 900000?

```python
def get_pdf_probability(dataset,startrange,endrange):
    from matplotlib import pyplot
    from scipy.stats import norm
    import seaborn as sns
    ax = sns.distplot(dataset,kde=True,kde_kws={'color':'blue'},color='Green')
    pyplot.axvline(startrange,color='Red')
    pyplot.axvline(endrange,color='Red')
    # generate a sample
    sample = dataset
    # calculate parameters
    sample_mean =sample.mean()
    sample_std = sample.std()
    print('Mean=%.3f, Standard Deviation=%.3f' % (sample_mean, sample_std))
    # define the distribution
    dist = norm(sample_mean, sample_std)
    
    # sample probabilities for a range of outcomes
    values = [value for value in range(startrange, endrange)]
    probabilities = [dist.pdf(value) for value in values]    
    prob=sum(probabilities)
    print("The area between range({},{}):{}".format(startrange,endrange,sum(probabilities)))
    return prob
get_pdf_probability(dataset["salary"],700000,900000)
```

## 12)Which parameter is highly correlated with salary?

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv('Placement_Data_Full_Class.csv')

# Filter only placed candidates and numerical columns
numerical_cols = ['ssc_p', 'hsc_p', 'degree_p', 'etest_p', 'mba_p', 'salary']
df_numerical = df[df['status'] == 'Placed'][numerical_cols].dropna()
```

```python
correlation_matrix = df_numerical.corr()
salary_correlations = correlation_matrix['salary'].sort_values(ascending=False)
print(salary_correlations)
```

## 13) plot  any useful graph and explain it.

```python
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0)
plt.title('Correlation Matrix with Salary')
plt.show()

# Plot top correlation
sns.regplot(x='mba_p', y='salary', data=df_numerical, scatter_kws={'alpha':0.3})
plt.title(f'MBA % vs Salary (Correlation: {salary_correlations["mba_p"]:.2f})')
plt.show()
```
