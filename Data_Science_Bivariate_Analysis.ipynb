#### Covariance**  
- **Definition**: Covariance measures the **direction of the linear relationship** (Difference) between two variables.  
- **Interpretation**:  
  - **Positive Covariance**: As \(X\) increases, \(Y\) tends to increase.  
  - **Negative Covariance**: As \(X\) increases, \(Y\) tends to decrease.  
  - **Zero Covariance**: No linear relationship.  
- **Limitation**:  
  - The magnitude is **not standardized**, so it’s hard to compare across datasets.

### Correlation (Pearson’s r)**  
- **Definition**: Correlation measures both the **direction and strength** of the linear relationship between two variables, standardized between 1 and -1 .  
- **Interpretation**:  
  - **+1**: Perfect positive linear relationship. (Types : Perfect, Low Degree, High Degree )
  - **-1**: Perfect negative linear relationship. (Types : Perfect, Low Degree, High Degree )
  - **0**: No linear relationship.  
- **Advantage**:  
  - Standardized, allowing comparison across different datasets.

### **VIF (Variance Inflation Factor) Explained**

#### **1. What is VIF?**
**VIF** measures **multicollinearity**—how much a predictor variable is linearly explained by other predictors in a regression model.  
- **High VIF** → The variable is highly correlated with others (redundant).  
- **Low VIF** → The variable provides unique information.  

#### **2. Interpretation**
- **VIF = 1**: No multicollinearity.  
- **1 < VIF < 5**: Moderate correlation (usually acceptable).  
- **VIF ≥ 5-10**: High multicollinearity (consider removing the variable).  

#### **. Why It Matters**
Multicollinearity inflates standard errors, making coefficient estimates unreliable in regression. VIF helps diagnose this issue.  

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
# Calculate VIF
def cal_vif(X):
    vif = pd.DataFrame()
    vif["Variable"] = X.columns
    vif["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
    
    return(vif)
```

### Hypothesis Testing Explained
Hypothesis testing is a statistical method used to determine whether there is enough evidence in a sample of data to infer that a certain condition holds true for the entire population.

### Make a Decision:

**Determine the P-value.

- Accept Null Hypothesis , if p-value > 0.05 or 5%
- Reject Null Hypothesis , if p-value < 0.05 or 5%
- Accept Alternative Hypothesis , if p-value < 0.05 or 5%
- Reject Alternative Hypothesis , if p-value > 0.05 or 5%

```python
from scipy.stats import ttest_rel
#dataset=dataset.dropna()
male = dataset[dataset['gender']=='M']['ssc_p']
male1 = dataset[dataset['gender']=='M']['hsc_p']
ttest_rel(male, male1)

#in this problem reject the Null , Accept the Alternative hypothesis. Becasue P Value is lower than 0.494 <0.5 or 5%.
```

### ANOVA TEST(How far rows and columns diffrenecs are went?)

ANOVA Test, or Analysis of Variance, is a statistical method used to test the differences between the means of two or more groups. Developed by Ronald Fisher in the early 20th century, ANOVA helps determine whether there are any statistically significant differences between the means of three or more independent groups.

- **The ANOVA analysis is a statistical relevance tool designed to evaluate whether or not the null hypothesis can be rejected while testing hypotheses. It is used to determine whether or not the means of three or more groups are equal.

- **Types of ANOVA  Formula
- **One-Way ANOVA
- **Two-Way ANOVA

### ANOVA Formula



- **F = ANOVA Coefficient
- **MSB = Mean of the total of squares between groupings
- **MSW = Mean total of squares within groupings
- **MSE = Mean sum of squares due to error
- **SST = total Sum of squares
- **p = Total number of populations
- **n = The total number of samples in a population
- **SSW = Sum of squares within the groups
- **SSB = Sum of squares between the groups
- **SSE = Sum of squares due to error
- **s = Standard deviation of the samples
- **N = Total number of observations

### One-Way Classification ANOVA
- **This test is used to see if there is a variation in the mean values of three or more groups. Such a test is used where the data set has only one independent variable. If the test statistic exceeds the critical value, the null hypothesis is rejected, and the averages of at least two different groups are significant statistically.

```python
from scipy.stats import f_oneway
import scipy.stats as stats
 
f_statistic, p_value = stats.f_oneway(dataset['ssc_p'], dataset['hsc_p'], dataset['degree_p'], )

print(f"F-statistic: {f_statistic}")
print(f"P-value: {p_value}")
```

- **H0 (null hypothesis): There is no significant difference between columns
- **H1 (alternative  hypothesis): There is significant difference between columns
- **Result : P(0.49)>0.05 or 5%, So H1 (alternative  hypothesis): There is significant difference between columns.

### Two-Way ANOVA
- **Two independent variables are used in the two-way ANOVA. As a result, it can be viewed as an extension of a one-way ANOVA in which only one variable influences the dependent variable. A two-way ANOVA test is used to determine the main effect of each independent variable and whether there is an interaction effect. Each factor is examined independently to determine the main effect, as in a one-way ANOVA. Furthermore, all components are analyzed at the same time to test the interaction impact.
